## 넘파이 실습

```python
import numpy as np

import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open('./assets/dasol.jpg') #이미지 불러오기(본인 이미지 경로에 맞게 수정하세요)
```

```python
img = np.array(img) #이미지를 numpy ndarray로 변환
```

```python
plt.imshow(img)
```

```python
img.shape # 1512X2016 행렬, 3차원(컬러)
```

```python
img.ndim
```

```python
plt.imshow(np.transpose(img, (1,0,2)))
```

```python
np.transpose(img, (1,0,2)).shape
#img의 2번째 축을 자신의 1번째 축으로 가짐 = 원본 이미지의 열이 자신의 행
#img의 1번째 축을 자신의 2번째 축으로 가짐 = 원본 이미지의 행이 자신의 열
```

```python
plt.imshow(img[:, :, :])
```

```python
plt.imshow(img[0:200, :, :])
```

```python
plt.imshow(img[:, 0:200, :])
```

```python
plt.imshow(img[:, :, 0]) #R채널
```

```python
plt.imshow(img[:, :, 1]) #G채널
```

```python
plt.imshow(img[:, :, 2]) #B채널
```

```python
plt.imshow(img[:, :, ::-1])
```

```python
plt.imshow(img[:, :, 2], cmap='gray') #B채널을 그레이스케일로 보기
```

```python
plt.imshow(img.mean(axis=2), cmap='gray') #RGB채널의 중앙값
```

```python
plt.imshow(img.mean(axis=0), cmap='gray')
```

```python
plt.imshow(img.mean(axis=1), cmap='gray')
```
